## Day 10 - Visualization

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/tmdb_movies_data.csv', parse_dates=['release_date', 'release_year'])
df.head()
```

```python
#listing names of columns
df.columns
```

## Movie with the most profit

```python
df['profits'] = df['revenue'] - df['budget']
df[df['profits'] == df['profits'].max()]
```

### >>Movie with the most profit is "Avatar"

## Movie with the greatest budget

```python
#movie with the greatest budget
df[df['budget'] == df['budget'].max()]
```

### >>Movie with the greatest budget is "The Warrior's Way"

## Movie with the greatest revenue

```python
#movie with the greatest revenue
df[df['revenue'] == df['revenue'].max()]
```

### >>Movie with greatest revenue is "Avatar"

## Month with highest number of movies

```python
#which month released the highest number of movies
df.index = df['release_date']
movie_count = df.resample('M').count()
max_num = movie_count['original_title'].max()
print(max_num)
movie_count[movie_count['original_title'] == max_num]
```

### >>September 2014 released the highest number of movies ie, 98 movies

## Month with the most Profit

```python
#which month made the most profit
new_df = df.resample('M').sum()
new_df[new_df['profits'] == new_df['profits'].max()]
```

### >>June 2015 made the most profits

## Finding average duration, budget, revenue and profits for 10 most popular movies

```python
# 10 most popular movies
df2 = df.copy()
df2.sort_values(['popularity'], ascending=False, inplace=True)
top10_max_pop = df2.head(10)
print('Average duration', top10_max_pop['runtime'].mean())
print('Average budget', top10_max_pop['budget'].mean())
print('Average revenue', top10_max_pop['revenue'].mean())
print('Average profits', top10_max_pop['profits'].mean())

top10_max_pop
```

## Graph of total profits made by all movies per year.

```python
profits_by_year = df.groupby(['release_year']).sum()
py = profits_by_year.reset_index()
py.head()
```

```python
plt.plot(py['release_year'], py['profits'])
plt.xlabel('Release Year of Movies')
plt.ylabel('Total Profits made by Movies')
plt.title('Calculating Profits made by movies in year released')
plt.show()
```

## Visualizing Vote Average

```python
plt.hist(df['vote_average'], bins=30)
plt.xlabel('Vote Average')
plt.ylabel('Num of Votes')
plt.title('Distribution of Vote Average over the years')
plt.show()
```

## Checking for correlation between budget and revenue

```python
# correlation between budget and revenue
plt.scatter(df['budget'], df['revenue'])
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.title('Budget Against Revenue')
plt.show()
```

### >>There is not much correlation between them

## Any other insights -- Finding correlation between popularity and profits

```python
plt.scatter(df['popularity'], df['profits'])
plt.xlabel('Popularity')
plt.ylabel('Profits')
plt.title('Popularity Against Profits')
plt.show()
```

### >> There is not much correlation between popularity and profits
